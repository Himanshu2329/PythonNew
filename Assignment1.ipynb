```python
# Print first 10 natural numbers using while loop
i=1
while i<=10:
    print(i)
    i+=1
```

```python
# Calculate sum of all numbers from 1 to a given number (1500)"

n=42
cnt=0
for i in range(1,n+1):
    cnt+=i
print(cnt)    
```

```python
# Print multiplication table of a given number (15 till 15 x 100)
n=15
for i in range(1,101):
    print(f"{n} x {1} = {i*n} ")
```

```python
# Print numbers in reverse order using a loop(100 to 5 )
for i in range(100,4,-1):
    print(i)
```

```python
# Print all prime numbers within a range without function
for i in range(1,101):
    if i>1:
        for j in range(2,i):
            if(i%j==0):
                break
            print(i)        
```
